```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("company_sales_data_matplotlib.csv")
total_profit_per_month = data.groupby('month_number')['total_profit'].sum()
plt.figure(figsize=(10, 6))
plt.plot(total_profit_per_month.index, total_profit_per_month.values, marker='o', linestyle='-')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.xticks(total_profit_per_month.index)
plt.grid(True)
plt.tight_layout()
plt.show()
```
