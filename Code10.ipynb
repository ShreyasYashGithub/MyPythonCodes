```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("company_sales_data_matplotlib.csv")

bathingsoap_sales = data['bathingsoap']
facewash_sales = data['facewash']
months = data['month_number']

fig, axs = plt.subplots(2, sharex=True, figsize=(10, 8))

axs[0].plot(months, bathingsoap_sales, color='b', marker='o')
axs[0].set_title('Bathing Soap Sales Over Months')
axs[0].set_ylabel('Units Sold')

axs[1].plot(months, facewash_sales, color='g', marker='o')
axs[1].set_title('Facewash Sales Over Months')
axs[1].set_xlabel('Month Number')
axs[1].set_ylabel('Units Sold')

plt.tight_layout()
plt.show()
```
