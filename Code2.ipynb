```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data_matplotlib.csv")

months = data['month_number']
facecream_sales = data['facecream']
facewash_sales = data['facewash']
toothpaste_sales = data['toothpaste']
bathingsoap_sales = data['bathingsoap']
shampoo_sales = data['shampoo']
moisturizer_sales = data['moisturizer']

plt.figure(figsize=(10, 6))
plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales, labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()

plt.show()
```
