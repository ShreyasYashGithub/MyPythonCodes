```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("company_sales_data_matplotlib.csv")
toothpaste_sales = data['toothpaste']
plt.figure(figsize=(10, 6))
plt.scatter(data['month_number'], toothpaste_sales, color='blue', marker='o')
plt.title('Toothpaste Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```
