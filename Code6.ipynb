```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("company_sales_data_matplotlib.csv")
facecream_sales = data['facecream']
facewash_sales = data['facewash']
months = data['month_number']
plt.figure(figsize=(10, 6))
plt.bar(months - 0.2, facecream_sales, width=0.4, color='b', align='center', label='Face Cream')
plt.bar(months + 0.2, facewash_sales, width=0.4, color='g', align='center', label='Face Wash')
plt.title('Face Cream and Face Wash Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(months)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
