```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("company_sales_data_matplotlib.csv")

bathingsoap_sales = data['bathingsoap']
months = data['month_number']

plt.figure(figsize=(10, 6))
plt.bar(months, bathingsoap_sales, color='c', align='center')
plt.title('Bathing Soap Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(months)
plt.grid(True)
plt.tight_layout()
plt.savefig('bathing_soap_sales.png')

plt.show()
```
