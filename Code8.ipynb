```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("company_sales_data_matplotlib.csv")
total_profit = data['total_profit']
plt.figure(figsize=(10, 6))
plt.hist(total_profit, bins=10, color='skyblue', edgecolor='black')
plt.title('Total Profit Distribution')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()

plt.show()
```
