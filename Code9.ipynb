```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("company_sales_data_matplotlib.csv")

total_sales_last_year = data.iloc[-12:, 1:7].sum()

plt.figure(figsize=(8, 8))
plt.pie(total_sales_last_year, labels=total_sales_last_year.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales Data for Last Year by Product')
plt.axis('equal') 

plt.show()
```
